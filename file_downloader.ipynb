```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd

page = requests.get("http://notelections.online/region/region/st-petersburg?action=show&root=1&tvd=27820001217417&vrn=27820001217413&region=78&global=&sub_region=78&prver=0&pronetvd=null&vibid=27820001217417&type=222")
soup = BeautifulSoup(page.text)

tables = soup.find_all('table')
links = tables[8].find_all('a', href = True)

tiks = []
urls = []
for link in links[:-1]:
    tiks.append(link.text)
    urls.append('http://notelections.online' + link['href'])
    
result = pd.DataFrame()

table = soup.find_all('table')[6].find('table')
columns = []
for col in table.find_all('tr'):
    column = col.find_all('td')[1].find('nobr').text
    columns.append(column)

ind = columns.index('\xa0')
columns.pop(ind)

ind = columns.index('Сумма')
columns.pop(ind)


for i in range (len(urls)):
    page_ = requests.get(urls[i])
    soup_ = BeautifulSoup(page_.text)
    div = soup_.find('div')
    uik = []
    for a in div.find('tr').find_all('a'):
        uik.append(a.text)
    tr = div.find_all('tr')
    tr.pop(0)
    tr.pop(11)    
    values = []
    for t in tr:
        val = []
        for b in t.find_all('b'):
            val.append(int(b.text))
        values.append(val)
    values = np.array(values)

    df = pd.DataFrame(columns = columns)
    
    for j in range (len(values)):
        d = {columns[j]:pd.Series(values[j], index = [uik])}
        df[columns[j]] = pd.DataFrame(d)[columns[j]]
    df['ТИК'] = tiks[i]

    result = result.append(df)

result.to_csv('data.csv')
```
